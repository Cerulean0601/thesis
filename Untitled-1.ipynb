```python
import math

filename = r"C:\Users\5F19 DS Lab\Desktop\HeJi\論文實驗\log.txt"
output = open(r"C:\Users\5F19 DS Lab\Desktop\HeJi\論文實驗\log2.txt", "w")

with open(filename, "r") as f:
    content = f.readline()
    content = content.split(",")[1:]

    count = 1
    for i in range(len(content)):
        if i%2 == 0:
            output.write("{0},{1},".format(count, content[i][1:]))
            count += 1
        else:
            mantissa = math.ceil(math.log(count, 10))
            output.write("{0}\n".format(content[i][1:-mantissa]))
```

```python
filename = r"C:\Users\5F19 DS Lab\Desktop\HeJi\論文實驗\log2.txt"
dp = [0]*13

with open(filename, "r") as f:
    for line in f:
        content = line.split(",")
        content[1] = int(content[1].split("=")[1])
        content[2] = float(content[2].split("=")[1])

        if content[1] != 0:
            dp[content[1]] = max(dp[content[1]-1], content[2])
        else:
            dp[0] = content[2]

with open(r"C:\Users\5F19 DS Lab\Desktop\HeJi\論文實驗\max.txt", "w") as f:
    for i in range(13):
        f.write(str(i) + "," + str(dp[i]) + "\n")
```
